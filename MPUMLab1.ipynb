Wygeneruj podobne obrazki (np.random.multivariate_normal, plt.scatter, plt.hist, 10k punktów (scatter), 100k (histogram), do załączonych (2 pkt):

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
mean = [2, 1]
cov = [[1.5, 0], 
       [0, 2]] 

x, y = np.random.multivariate_normal(mean, cov, 10000).T
plt.plot(x, y, 'o')
plt.show()
```

```python
mean = [1, 2]
x = 0.00001
cov = [[1, x], 
       [  x, 0.0001]] 

x, y = np.random.multivariate_normal(mean, cov, 10000).T
plt.plot(x, y, 'o')
plt.show()
```

```python
mean = [0, 2]
x = 6
cov = [[6, x], 
       [  x, 6]] 

x, y = np.random.multivariate_normal(mean, cov, 10000).T
plt.plot(x, y, 'o')
plt.show()
```

```python
s = np.random.normal(0, 1, 100000)
n, bins, p = plt.hist(s, 'auto',  normed=True)

density = (1 / np.sqrt(2 * np.pi)) * np.exp(- (bins * bins) / 2)
plt.plot(bins, density)

plt.show()
```

```python
import scipy
from scipy import linalg

def gaussN(u, sigma):
    dim = len(u)
    if sigma.shape != (dim, dim):
        raise "Zły rozmiar macierzy kowariancji"
        Z = np.random.normal(0, 1, size = dim)
    B = scipy.linalg.sqrtm(sigma)
    return Z.dot(B) + u
    
```

```python
cov = np.array([[3, 0], [0, 4]])
u = np.array([2, 4])


points = []

for i in range(0, 10000):
    points.append(gaussN(u, cov))

points = np.array(points)
plt.plot(points[:,0], points[:,1], 'o')
plt.show()
    
```
